## Installing Python and Anaconda on Windows

Before getting started with the installation, let’s learn a bit more about what Anaconda exactly is. Python is the programming language which will be installed on the machine and on top of that different independent development environments (IDEs) and packages can be installed. Python on its own is not going to be very useful unless an IDE is installed. This is where Anaconda comes in the picture.

Anaconda installs IDEs and several important packages like NumPy, Pandas, and so on.


**Step 1: Download the installer from the link below.**

[Downloads - Anaconda](https://www.anaconda.com/download/)


**Step 2: Scroll down and select the operating system you are using.  I have selected Windows.**

![download](images/00/py0_1.png)

**Step 3: From the Python 3.7 section, choose from 32bit / 64bit options. To determine the correct option, in File Explorer:**

* Right click This PC 
* Select Properties.

![properties](images/00/py0.png)

* Select the appropriate download

**Step 4: Go through the installation procedure with the installer.**

* Open the installer. Click Next.

![install1](images/00/py1.png)

* Click - I Agree.

![install2](images/00/py2.png)

* Select Just Me. Click Next

![install3](images/00/py3.png)

* Click Next

![install4](images/00/py4.png)

* Select Register Anaconda … Click Install

![install5](images/00/py5.png)

* Click Next a few times until the Finish button appears.  Click Finish

**Step 6: After the installation is complete, open the Anaconda Prompt from the Start menu.**

At the prompt type 
	
    conda update conda
